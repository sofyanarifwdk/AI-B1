```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1️⃣ Data Awal: Nilai Hasil Evaluasi Kinerja Karyawan
# ------------------------------------------------------------
data = {
    'Nama': ['Deni Suryono', 'Andri Priyono', 'Syamsul Ma\'arif', 'Budi Purnama', 'Cipta Ningsih'],
    'Kehadiran': [85, 95, 75, 80, 70],
    'Kualitas_Kerja': [70, 85, 50, 80, 60],
    'Kreatifitas': [64, 78, 63, 72, 57],
    'Technical_Skills': [80, 82, 53, 75, 67],
    'Communication_Skills': [64, 81, 50, 63, 55],
    'Attitude': [65, 75, 52, 60, 62]
}

df = pd.DataFrame(data)
display(df)
```

```python
# 2️⃣ Fungsi Keanggotaan (Buruk, Cukup, Baik)
# ------------------------------------------------------------

def linear_up(x, a, b):
    if x <= a:
        return 0
    elif x >= b:
        return 1
    else:
        return (x - a) / (b - a)

def linear_down(x, a, b):
    if x <= a:
        return 1
    elif x >= b:
        return 0
    else:
        return (b - x) / (b - a)

def triangular(x, a, b, c):
    if x <= a or x >= c:
        return 0
    elif a < x < b:
        return (x - a) / (b - a)
    elif b <= x < c:
        return (c - x) / (c - b)
    else:
        return 0
```

```python
# 3️⃣ Visualisasi Fungsi Kehadiran (semesta 0–100)
# ------------------------------------------------------------

x = np.linspace(0, 100, 200)
buruk = [linear_down(i, 75, 85) for i in x]
cukup = [triangular(i, 75, 85, 95) for i in x]
baik = [linear_up(i, 85, 95) for i in x]

plt.figure(figsize=(8,4))
plt.plot(x, buruk, label='Buruk', color='red')
plt.plot(x, cukup, label='Cukup', color='orange')
plt.plot(x, baik, label='Baik', color='green')
plt.title('Fungsi Keanggotaan (Buruk–Cukup–Baik)')
plt.xlabel('Nilai (0–100)')
plt.ylabel('Derajat Keanggotaan')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ------------------------------------------------------------
# 2️⃣ Data fuzzifikasi dari Tabel 3–8 jurnal
# ------------------------------------------------------------
data_fuzzy = {
    'Nama': ['Deni Suryono', 'Andri Priyono', "Syamsul Ma'arif", 'Budi Purnama', 'Cipta Ningsih'],
    # Kehadiran
    'Kehadiran_Buruk': [0, 0, 1, 0.5, 1],
    'Kehadiran_Cukup': [1, 0, 0, 0.5, 0],
    'Kehadiran_Baik':  [0, 1, 0, 0, 0],
    # Kualitas Kerja
    'Kualitas_Kerja_Buruk': [0, 0, 1, 0, 0],
    'Kualitas_Kerja_Cukup': [0, 0, 0, 0, 1],
    'Kualitas_Kerja_Baik':  [0.5, 1, 0, 1, 0],
    # Kreatifitas
    'Kreatifitas_Buruk': [0, 0, 0, 0, 0.3],
    'Kreatifitas_Cukup': [0.6, 0, 0.7, 0, 0.7],
    'Kreatifitas_Baik':  [0.4, 1, 0.3, 1, 0],
    # Technical Skills
    'Technical_Skills_Buruk': [0, 0, 1, 0, 0],
    'Technical_Skills_Cukup': [0, 0, 0, 0.3, 0.9],
    'Technical_Skills_Baik':  [1, 1, 0, 0.7, 0.1],
    # Communication Skills
    'Communication_Skills_Buruk': [0.1, 0, 1, 0.1, 0.7],
    'Communication_Skills_Cukup': [0.9, 0, 0, 0.9, 0.3],
    'Communication_Skills_Baik':  [0, 1, 0, 0, 0],
    # Attitude
    'Attitude_Buruk': [0, 0, 0.8, 0, 0],
    'Attitude_Cukup': [0.5, 0, 0.2, 1, 0.8],
    'Attitude_Baik':  [0.5, 1, 0, 0, 0.2]
}
df_fuzzy = pd.DataFrame(data_fuzzy)
display(df_fuzzy)
```

```python
# ------------------------------------------------------------
# 3️⃣ Fuzzifikasi Query: semua variabel "Baik"
# ------------------------------------------------------------
baik_cols = [c for c in df_fuzzy.columns if c.endswith('_Baik')]
hasil_df = df_fuzzy[['Nama'] + baik_cols].copy()
hasil_df['Hasil_Evaluasi'] = hasil_df[baik_cols].mean(axis=1)
hasil_df = hasil_df.sort_values(by='Hasil_Evaluasi', ascending=False).reset_index(drop=True)
display(hasil_df)

```

```python
# ------------------------------------------------------------
# 4️⃣ Hasil akhir dan visualisasi
# ------------------------------------------------------------
print("🏆 HASIL EVALUASI FUZZY TAHANI (Sesuai Jurnal)\n")
for i, r in hasil_df.iterrows():
    print(f"{i+1}. {r['Nama']} → Nilai: {round(r['Hasil_Evaluasi'],2)}")

print("\nKaryawan terbaik:")
print(f"👉 {hasil_df.iloc[0]['Nama']} (nilai: {round(hasil_df.iloc[0]['Hasil_Evaluasi'],2)})")

plt.figure(figsize=(7,4))
plt.bar(hasil_df['Nama'], hasil_df['Hasil_Evaluasi'], color='lightblue')
plt.title('Nilai Fuzzy Tahani (Sesuai Jurnal)')
plt.ylabel('Rata-rata Keanggotaan "Baik"')
plt.xticks(rotation=30)
plt.grid(axis='y', alpha=0.6)
plt.show()
```
